```python
# Importing the neccessary libraries
import pandas as pd
import time
import requests
from  bs4 import BeautifulSoup
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException,ElementNotVisibleException
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

## Scraping data from Yatra.com

```python
#path for chromedriver
driver = webdriver.Chrome('chromedriver.exe')

# to open the webpage
url = "https://www.yatra.com/"
driver.get(url)
driver.maximize_window()
time.sleep(2)
```

```python
driver.find_element(by=By.XPATH, value="//input[@class='js_ripple search-btn ']").click()
time.sleep(3)
```

```python
#empty lists to store data
airline_name=[]
flight_code=[]
source = []
destination = []
departure_time = []
arrival_time = []
duration=[]
total_stops = []
date_of_journey=[]
meal=[]
price=[]
location=[]
```

```python
#source
source = driver.find_element(by=By.XPATH, value="//div[@class='input-holder pb-2 bdr-btm']/input")
time.sleep(2)
source.clear() 
source.send_keys('New Delhi')
time.sleep(2)
            
#destination
destination = driver.find_element(by=By.XPATH, value="//div[@class='input-holder  bdr-btm pb-2']/input")
time.sleep(2)
destination.click()
destination.clear()
time.sleep(2)
destination.send_keys('Mumbai')
time.sleep(2)
    
#click on search button
try:    
    srch_btn = driver.find_element(by=By.XPATH, value="//button[@class='fs-14 btn-submit cursor-pointer bold']")
    wait = WebDriverWait(driver, 10)
    wait.until(EC.visibility_of(srch_btn))
    srch_btn.click()
except NoSuchElementException:
    pass
time.sleep(1)
            
#scrolling the page
for _ in range(15):
    time.sleep(1)
    driver.execute_script("window.scrollBy(0,5000)")
    time.sleep(3)
            
#fetching name of airline
try:
    for i in driver.find_elements(by=By.XPATH, value="//div[@class='fs-13 airline-name no-pad col-8']/span"):
        airline_name.append(i.text)
except NoSuchElementException:
    pass
#fetching flight code
try:
    for i in driver.find_elements(by=By.XPATH, value="//p[@class='normal fs-11 font-lightestgrey no-wrap fl-no']"):
        flight_code.append(i.text)
except NoSuchElementException:
    pass
#fetching source and destination name 
try:
    for i in driver.find_elements(by=By.XPATH, value="//p[@class='fs-10 font-lightgrey no-wrap city ellipsis']"):
        location.append(i.text)
except NoSuchElementException:
    pass
#fetching time of departure
try:
    for i in driver.find_elements(by=By.XPATH, value="//div[@class='i-b pr']"):
        departure_time.append(i.text)
except NoSuchElementException:
    pass
#fetching time of arrival
try:
    for i in driver.find_elements(by=By.XPATH, value="//div[@class='i-b pdd-0 text-left atime col-5']//p[1]"):
        arrival_time.append(i.get_attribute("innerHTML").splitlines()[0][0:5])
except NoSuchElementException:
    pass
#fetching duration
try:
    for i in driver.find_elements(by=By.XPATH, value="//div[@class='stop-cont pl-13']/p"):
        duration.append(i.text)
except NoSuchElementException:
    pass
#fetching number of stops
try:
    for i in driver.find_elements(by=By.XPATH, value="//div[@class=' font-lightgrey fs-10 tipsy i-b fs-10']/span[1]"):
        total_stops.append(i.text)
except NoSuchElementException:
    pass
#fetching availability of meal
try:
    for i in driver.find_elements(by=By.XPATH, value="//div[@class='features pull-right fs-12 flex']/div[1]"):
        meal.append(i.text)
except NoSuchElementException:
    pass
#fetching prices
try:
    for i in driver.find_elements(by=By.XPATH, value="//div[@class='i-b tipsy fare-summary-tooltip fs-18']"):
        price.append(i.text)
except NoSuchElementException:
    pass
            
```
